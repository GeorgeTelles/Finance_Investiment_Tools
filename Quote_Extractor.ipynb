# Quote Extractor

## Description

The `Quote Extractor` project aims to obtain and transform exchange rates for various currencies into the Brazilian Real (BRL). Utilizing the [AwesomeAPI](https://economia.awesomeapi.com.br/), this project retrieves the current exchange rates for major currencies and cryptocurrencies such as USD (U.S. Dollar), EUR (Euro), and BTC (Bitcoin), and converts them into BRL.

## Features

- **Exchange Rate Retrieval**: Fetches current exchange rates for USD, EUR, and BTC in relation to the Brazilian Real (BRL) from the AwesomeAPI.
- **Data Transformation**: Converts the exchange rates into a Pandas DataFrame for easy viewing and analysis.
- **Visualization**: Presents the exchange rates in a clear and organized format, showing the respective currencies and their BRL values.

## Code

The main code performs the following steps:

1. **API Request**: Sends a GET request to the AwesomeAPI to retrieve the exchange rates for USD, EUR, and BTC in BRL.
2. **Data Processing**: Parses the JSON response from the API and extracts the relevant exchange rates.
3. **DataFrame Creation**: Organizes the exchange rates into a Pandas DataFrame for easy visualization and manipulation.



```python
import requests
import pandas as pd
```

```python
def pegar_cotacoes():
    requisicao = requests.get("https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL")

    requisicao_dic = requisicao.json()

    cotacao_dolar = requisicao_dic['USDBRL']['bid']
    cotacao_euro = requisicao_dic['EURBRL']['bid']
    cotacao_btc = requisicao_dic['BTCBRL']['bid']

    cotacoes = pd.DataFrame({'Moeda': ['Dólar/BRL', 'Euro/BRL', 'BTC/BRL'],
                            'Cotação': [cotacao_dolar, cotacao_euro, cotacao_btc]})

    return cotacoes
```

```python
pegar_cotacoes()
```
